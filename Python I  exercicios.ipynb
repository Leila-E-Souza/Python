### Exercício 3)
Dada a lista `meus_numeros` abaixo, crie uma nova lista chamada `meus_numeros3` contendo apenas elementos de `meus_numeros` que são multiplos de 3.

__Dica:__ Utilize o operador binário <font color='blue'>'%'</font> que retorna o resto da divisão de um número por outro. Por exemplo:
```python
print(22%6)
```
resulta no valor 4

````python
#aqui ja temos a lista principal
meus_numeros = [1, 56, 342, 12, 781, 23, 43, 45, 123, 567]

meus_numeros3 = [] #declarando a nova variavel
for i in meus_numeros: #condicao que diz que I representa a lista
    if i%3 == 0: #se o resto da divisao por 3 for = 0, entao adiciona o elemento a nova lista
        meus_numeros3.append(i) #nova lista com elementos multiplos de 3
#imprime        
print(meus_numeros3)
````

### Exercício 5) 
Considere a lista `minhas_notas` definida abaixo, onde cada elemento é uma lista que representa notas dadas a um produto. Calcule a média de notas de cada produto, salvando o resultado em uma nova lista chamada `notas_medias`.

__Dica:__ Utilize o comando <font color='blue'>'sum'</font> do Python para somar os valores em uma lista. Por exemplo:
```python
ls = [1,1,1,1,1]
print(sum(ls))
```

````python
minhas_notas = [[2.3, 4.5, 1.2], # Notas do produto 1
                [5.4, 3.6, 7.9, 6.2], # Notas do produto 2
                [8.1, 8.9, 6.5, 7.9, 9.4], # ...
                [9.8, 7.6],
                [3.2, 4.5, 1.6, 4.8]]
notas_medias = [] #lista da media dos produtos
for i in minhas_notas: #funcao I na lista minnhas notas
    notas_medias.append(sum(i)/len(i)) #nova lista com a divisao da soma dos elementos de I
    
#Para lembrar: (sum(i)/len(i) aqui fazemos a soma dos elementos do prod 1 e dividimos pelo tamanho do prod

#imprime
print("media da nota dos produtos eh:", notas_medias)
````

### Exercício 6) 
Considerando a lista `nomes` a abaixo:

```python
print('us' in 'Gustavo')
```
imprime como resultado:
```python
True
```

````python
nomes = ['Matheus', 'Gabriel', 'Gustavo', 'Hugo', 'Ricardo', 'Rui', 'Guilherme', 'Luiz']

nomes_o = [] #criando a lista nova com os elementos q tem a letra 'O'
for n in nomes: #passando n como variavel na lista nomes 
    if 'o' in n: #condicao que 'IN' identifica se a letra O esta em algum dos elementos de N (nomes)
        nomes_o.append(n) #lista com os elementos que tem a letra 'o'
#imprime
print(nomes_o)
````

### Exercício 7) 
Ainda utilizando a lista `nomes` definida a cima:
- crie uma nova lista chamada `concatenados` com os elementos da lista `nomes`, porém, elementos começando com a mesma letra devem ser concatenados em um único elemento, ou seja, a lista resultante deve ser: 
```python
['Matheus', 'GabrielGustavoGuilherme', 'Hugo', 'RicardoRui', 'Luiz']
```
__Dica:__ Crie um dicionario onde a chave corresponde uma letra e o valor a uma string que pode ser concatenada caso o valor não seja vazio. Você pode utilizar o comando <font color='blue'>'in'</font> para verificar se uma chave existe em um dicionário. Por exemplo, o código:
```python
dc = {'c1': 0, 'c2':1, 'c3':2}
print('c2' in dc)
```
imprime como resultado:
```python
True
```

````python
nomes = ['Matheus', 'Gabriel', 'Gustavo', 'Hugo', 'Ricardo', 'Rui', 'Guilherme', 'Luiz']

dic_nomes = {} #cria o dicionario com os nomes
for n in nomes:
    if n[0] in dic_nomes:
        dic_nomes[n[0]] = dic_nomes[n[0]]+n
    else:
        dic_nomes[n[0]] = n
 
concatenados = list(dic_nomes.values()) # converte os valores do dicionario em uma lista
print(concatenados)
````

### Exercício 10) 
Escreva um código para contar o número de ocorrências de cada letra na string fornecida e imprima os resultados.

__Dica:__ Utilize um dicionário.

````python
minha_string = "scientia vinces"

docrr = {} #definindo o dicionario
for c in minha_string: #definindo C na lista da palavra
    if c in docrr: 
        docrr[c] += 1
    else:
        docrr[c] = 1
    
for vc in docrr.items():
    print(vc[0],'=',vc[1])
````
